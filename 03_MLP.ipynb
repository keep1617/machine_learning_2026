```python

import numpy as np
from load import data_loader
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
def load_all_npz(folder_path):
    ## npzfile 불러오기
    folder = Path(folder_path)
    npz_files = list(folder.glob("*.npz"))
    return npz_files
```

```python

### load npz
npz_file_list = load_all_npz("./Tactile_sensor_test_data")
print(npz_file_list)

x_data = np.empty((0, 5))
y_data = np.empty((0, 1))


print(x_data.shape)
for file in npz_file_list:
    path = "./Tactile_sensor_test_data" + f"/{file.name}"
    
    file_force_npz = path.split("_")[-1]

    force = int(file_force_npz.replace("N.npz",""))


    x_tactile = data_loader(path) ## -3으로 
    x_top_3 = np.partition(x_tactile, -5, axis=1)[:,-5:]

    ## parse force by name



    y_label = np.full((x_top_3.shape[0],1),force)
    x_data = np.vstack([x_data,x_top_3])
    y_data = np.vstack([y_data,y_label])
```
